# QUS 1

```python
#import the necessary libraries:
import selenium
import pandas as pd
from selenium import webdriver
import warnings
warnings.filterwarnings('ignore')
from selenium.webdriver.common.by import By
import time
```

```python
#set the selenium webdriver:

driver = webdriver.Chrome()
```

```python
#open the webpage:

driver.get('https://www.shine.com/')
```

```python
#enter the job title and location:

job_title = driver.find_element(By.NAME,"id_q")
job_title.send_keys('Data Analyst')

location = driver.find_element(By.NAME,"id_loc")
location.send_keys('Bangalore')
```

```python
#click the search button:

search = driver.find_element(By.XPATH,"/html/body/div/div[1]/div/div/div[1]/div[1]/div/div[2]/div/div/form/div/div[2]/div/button")
search.click()
```

```python
job_title =[]
job_location =[]
company_name =[]
experience_requried =[]
```

```python
#scrape the data:

title_tags = driver.find_elements(By.XPATH,'//div[@class="jobCard_jobCard_features__wJid6"]')
for i in title_tags:
    title = i.text
    job_title.append(title)
    

location_tags = driver.find_elements(By.XPATH,'//div[@class=" jobCard_jobCard_lists_item__YxRkV jobCard_locationIcon__zrWt2"]')
for i in location_tags:
    location = i.text
    job_location.append(location)    

company_tags = driver.find_elements(By.XPATH,'//div[@class="jobCard_jobCard_cName__mYnow"]')
for i in company_tags:
    company = i.text
    company_name.append(company)
    

experience_tags = driver.find_elements(By.XPATH,'//div[@class=" jobCard_jobCard_lists_item__YxRkV jobCard_jobIcon__3FB1t"]')
for i in experience_tags:
    exp = i.text
    experience_requried.append(exp)
    
```

```python
print(len(job_title),len(job_location),len(company_name),len(experience_requried))
```

```python
#creat a dataframe:

import pandas as pd
df = pd.DataFrame({'Title':job_title[:10], 'Location':job_location[:10], 'Company Name':company_name[:10],'Experience':experience_requried[:10]})
df
```

```python
#close the driver

driver.quit()
```
